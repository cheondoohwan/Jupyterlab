```python
from operator import add
```

```python
counts = lines.flatMap(lambda x: x.split(' ')) \
    .map(lambda x: (x, 1)).reduceByKey(add)
```

```python
output = counts.collect()
```

```python
for (word, count) in output:
    print(f'({word}, {count})')
```
